```python
#Basic Mathematics
df['Revenue'] = df['Quantity'] * df['Revenue per product']
df['Cost'] = df['Quantity'] * df['Cost per product']
df['Profit'] = df['Revenue'] - df['Cost']
df
```

```python
#SORTING A TABLE 
#df.sort_values('Countries')
#df.sort_values('Countries', ascending = False)
df.sort_values(['Countries', 'Quantity'], ascending = [1,0])
```

```python
df["Month"] = df["Order Date"].apply(lambda x : x. strftime('%B'))
df["Year"] = df["Order Date"].apply(lambda x : x. strftime('%Y'))
df
```

```python
#To filter out specific info
df.loc[df['Countries'] == "Nigeria"]
```

```python
#Filters by a single products
df.query("Products == 'Pin Charger'")
```

```python
#Filters other products except pin charger
df.query("Products ! = 'Pin Charger'")
```

```python
#Two types of products(Type A Charger and Pin Charger)
df.query("Products in ('Type A Charger', 'Pin Charger')")
```

```python
#Profit Greater than N-Value
df.query("`Profit` >= 2000")
```

```python
#Filters between different values
df.query("`Profit` >= 2000 and `Profit` <=7000")
```

```python
#Filters between Dates
df.query("`Order Date` >= '2022-02-01' and `Order Date` <= '2022-05-31'")
```

```python
#Filters in Ember months
df[df["Month"].str.contains("ber")]
```

```python
#Filtering based on multiple conditions(Countries and Products)
df.query("`Countries` == 'Nigeria' and `Products` == 'Type A Charger'")
```

```python
#Filtering based on multiple conditions(Revenue)
df.query("`Countries` == 'Nigeria' and `Revenue` >= 2000")
```

```python
#Filtering Cost, Products andd Profit
#df.query("`Products` == 'Type A Charger' and `Cost`>=1000 and `Profit` <3000")
df.query("`Products` in('Type A Charger', 'Pin Charger') and `Cost` >=1000 and `Profit` <3000")
```

```python
#Filtering profits between dates that is greater than N-Value
df.query("`Profit` >=3000 and `Order Date` >= '2022-02-01' and `Order Date` <= '2022-05-31'")
```

```python
#Filtering between Ember months
#df.query("`Profit` > 3000 and `Month` in('September', 'October', 'November', 'December')")
df.query("`Profit` > 3000 and `Order Date` > '2022-08-31'")
```

```python
#PIVOT TABLE - SUMMARY OF THE PRODUCTS
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df.pivot_table(index="Products", values=numeric_cols, aggfunc="sum")

df.pivot_table(index="Products", values=["Quantity", "Cost per product", "Revenue per product", "Revenue", "Profit", "Cost"], aggfunc="sum")
```

```python
#Summary of Profits by Products
df.pivot_table(index = "Products",
               values = "Profit",
               aggfunc= "sum")
```

```python
#Summary of Profit, Revenue by Products
df.pivot_table(index = "Products",
               values = ["Profit", "Revenue"],
               aggfunc= "sum")
```

```python
#Summary of Profits by Products and Countries
df.pivot_table(index = ["Products", "Countries"],
               values = "Profit",
               aggfunc= "sum")


df.pivot_table(index = "Products",
               columns = "Countries",
               values = "Profit",
               aggfunc= "sum")
```

```python
#Summary of Profits, Revenue by Products and Countries
df.pivot_table(index = "Products",
               columns = "Countries",
               values = ["Profit", "Revenue"],
               aggfunc= "sum")
```

```python
#Summary of Profits by Months and Year
df.pivot_table(index = "Month",
               columns = "Year",
               values = "Profit",
               aggfunc= "sum")
```

```python
#Sort Month in Descending order
dfpivot = df.pivot_table(index = "Month",
               columns = "Year",
               values = "Profit",
               aggfunc= "sum")
dfpivot.sort_values(by=['Month'], ascending = True)
```

```python
#Sort Month in order from January - December
dfpivot = df.pivot_table(index = "Month",
               columns = "Year",
               values = "Profit",
               aggfunc= "sum")
month_dict ={'January':1, 'February':2, 'March':3,'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
dfpivot.sort_values('Month', key = lambda x : x.apply(lambda x : month_dict[x]))
```

```python
#Column Chart of Profit by Products without Labelling
barchart = df.pivot_table(index = "Products",
                         values = "Profit",
                         aggfunc = "sum")
barchart.plot(kind = "bar")
```

```python
#Column Chart of Profit by Products with Labelling
from matplotlib import pyplot as plt
columnchart = df.pivot_table(index = "Products",
                         values = "Profit",
                         aggfunc = "sum")
columnchart.plot(kind = "bar")
plt.title("Profit by Products")
plt.xlabel("Profit")
plt.ylabel("Products")
```

```python
#Bar Chart of Profit by Products with Labelling
from matplotlib import pyplot as plt
barchart = df.pivot_table(index = "Products",
                         values = "Profit",
                         aggfunc = "sum")
barchart.plot(kind = "barh")
plt.title("Profit by Products")
plt.xlabel("Profit")
plt.ylabel("Products")
```

```python
#Pie Chart of Revenue by Countries with Labelling
from matplotlib import pyplot as plt
piechart = df.pivot_table(index = "Countries",
                         values = "Revenue",
                         aggfunc = "sum")
piechart.plot(kind = "pie", y = 'Revenue', autopct = '%1.1f%%')
plt.title("Revenue by Countries")
```

```python
#Sorted Column Chart
from matplotlib import pyplot as plt
dfpivot = df.pivot_table(index = "Month",
                         values = "Profit",
                         aggfunc = "sum")
month_dict ={'January':1, 'February':2, 'March':3,'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
dftable = dfpivot.sort_values('Month', key = lambda x : x.apply(lambda x : month_dict[x]))
dftable.plot(kind = "bar")

plt.title("Profit by Month")
plt.xlabel("Profit")
plt.ylabel("Month")
```

```python
#Line Chart
from matplotlib import pyplot as plt
dfpivot = df.pivot_table(index = "Month",
                         values = "Profit",
                         aggfunc = "sum")
month_dict ={'January':1, 'February':2, 'March':3,'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
dftable = dfpivot.sort_values('Month', key = lambda x : x.apply(lambda x : month_dict[x]))
dftable.plot(kind = "line")

plt.title("Profit by Month")
plt.xlabel("Profit")
plt.ylabel("Month")
```

```python
#Area Chart
from matplotlib import pyplot as plt
dfpivot = df.pivot_table(index = "Month",
                         values = "Profit",
                         aggfunc = "sum")
month_dict ={'January':1, 'February':2, 'March':3,'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
dftable = dfpivot.sort_values('Month', key = lambda x : x.apply(lambda x : month_dict[x]))
dftable.plot.area()

plt.title("Profit by Month")
plt.xlabel("Profit")
plt.ylabel("Month")
```
